```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r"D:\Datasets\datasets\wisc_bc_data.csv")
df
```

```python
df.columns
```

```python
x = df.drop(['id', 'diagnosis'], axis = 1)
y = df['diagnosis']
```

```python
x.describe()
```

```python
# feature scaling
```

```python
from sklearn.preprocessing import MinMaxScaler
sca = MinMaxScaler()
x_scaled = sca.fit_transform(x)
```

```python
# cross validation
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x_scaled, y,
                                                    random_state= 0,
                                                    test_size= 0.25,
                                                    stratify= y)
```

```python
x_train.shape, y_train.shape
```

```python
# knn model

from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors= 18)
knn.fit(x_train, y_train)
```

```python
y_pred = knn.predict(x_test)
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

#### Hyperparameter Tuning

```python
# create a parameter grid
```

```python
param_grid = {
    'n_neighbors': [4, 5, 7, 9, 13, 17, 18, 20],
    'weights': ['uniform', 'distance'],
    'algorithm': ['auto', 'ball_tree', 'kd_tree', 'brute']
}
```

```python
# grid search algorithm
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid = GridSearchCV(estimator= KNeighborsClassifier(),
                    param_grid= param_grid,
                    scoring= 'accuracy',
                    cv = 5)
```

```python
grid.fit(x_scaled, y)
```

```python
grid.best_estimator_
```

```python
y_pred = grid.predict(x_test)
```

```python
accuracy_score(y_test, y_pred)
```

```python
# Randomised Search CV
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
rgrid = RandomizedSearchCV(estimator= KNeighborsClassifier(),
                           param_distributions= param_grid,
                           scoring= 'accuracy',
                           cv = 5,
                           n_iter= 10)
```

```python
rgrid.fit(x_scaled, y)
```

```python
rgrid.best_estimator_
```
