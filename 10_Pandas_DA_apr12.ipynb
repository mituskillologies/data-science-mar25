```python
df[df["marks"] > 60]
# students belonging to BE
```

```python
df[df["class"] == "BE"]
# marks > 60 and class = BE
```

```python
df[(df["marks"] > 60) & (df["class"] == "BE")]
```

```python
df[(df["marks"] > 60) | (df["class"] == "BE")]
```
