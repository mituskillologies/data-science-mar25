```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Social_Network_Ads.csv")
df
```

```python
# separate the columns in 2d format
# plot the scatterplot between age and estimated salary
```

```python
x = df[["Age", "EstimatedSalary"]]
```

```python
x.describe();
```

```python
plt.figure(figsize=(16, 9))
plt.title("age vs est sal")
plt.xlabel("est sal")
plt.ylabel("age")
plt.scatter(df["EstimatedSalary"], df["Age"], c = df["Purchased"]);
plt.grid()
```

#### Normalization

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
x_scaled = scaler.fit_transform(x)
```

```python
x_scaled;
```

```python
x_scaled[:, 0].max(), x_scaled[:, 0].min()
```

```python
x_scaled[:, 1].max(), x_scaled[:, 1].min()
```

```python
pd.DataFrame(x_scaled).describe()
```

```python
plt.figure(figsize=(16, 9))
plt.title("age vs est sal")
plt.xlabel("est sal")
plt.ylabel("age")
plt.scatter(x_scaled[:, 1], x_scaled[:, 0], c = df["Purchased"]);
plt.grid()
```

#### Standard scaler

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
x_scaled = scaler.fit_transform(x)
```

```python
pd.DataFrame(x_scaled).describe()
```

```python
plt.figure(figsize=(16, 9))
plt.title("age vs est sal")
plt.xlabel("est sal")
plt.ylabel("age")
plt.scatter(x_scaled[:, 1], x_scaled[:, 0], c = df["Purchased"]);
plt.grid()
```
