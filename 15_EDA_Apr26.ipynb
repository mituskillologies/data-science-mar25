### Data cleaning

```python
df.columns = df.columns.str.strip()
```

```python
df.isnull().sum()
```

```python
null_col = (df.isnull().sum() * 100) / len(df.index)
null_col;
```

```python
null_col_del = null_col[null_col > 30.0]
null_col_del
```

```python
col_drop = list(null_col_del.index)
col_drop;
```

```python
df.drop(col_drop, axis = 1, inplace = True)
```
