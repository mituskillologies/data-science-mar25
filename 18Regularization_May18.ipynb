```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"D:\Datasets\datasets\BostonHousing.csv")
df.head()
```

```python
df.isnull().sum().sum()
```

```python
# separate the data
```

```python
x = df.drop('medv', axis = 1)
y = df['medv']
```

```python
# feature scaling

from sklearn.preprocessing import StandardScaler
sca = StandardScaler()
x = sca.fit_transform(x)
```

#### Linear Regression

```python
# cross validation
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y,
                                                   random_state= 0,
                                                   test_size= 0.25)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
print(f"training accuracy of linear regression => {lr.score(x_train, y_train)}")
print(f"testing accuracy of linear regression => {lr.score(x_test, y_test)}")
```

#### Polynomial Regression

```python
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(degree = 2)

x_poly = poly.fit_transform(x)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_poly, y,
                                                   random_state= 0,
                                                   test_size= 0.25)
```

```python
lr_poly = LinearRegression()
lr_poly.fit(x_train, y_train)
```

```python
print(f"training accuracy of polynomial regression => {lr_poly.score(x_train, y_train)}")
print(f"testing accuracy of polynomial regression => {lr_poly.score(x_test, y_test)}")
```

#### Ridge Regression

```python
from sklearn.linear_model import Ridge
```

```python
ridge = Ridge(alpha = 18)
```

```python
ridge.fit(x_train, y_train)
print(f"training accuracy of ridge regression => {ridge.score(x_train, y_train)}")
print(f"testing accuracy of ridge regression => {ridge.score(x_test, y_test)}")
```

#### Lasso Regression

```python
from sklearn.linear_model import Lasso
```

```python
lasso = Lasso(alpha= 0.1)
```

```python
lasso.fit(x_train, y_train)
print(f"training accuracy of lasso regression => {lasso.score(x_train, y_train)}")
print(f"testing accuracy of lasso regression => {lasso.score(x_test, y_test)}")
```

#### Elastic net

```python
from sklearn.linear_model import ElasticNet
```

```python
net = ElasticNet(alpha= 0.1, l1_ratio= 0.6)
```

```python
net.fit(x_train, y_train)
print(f"training accuracy of elastic net regression => {net.score(x_train, y_train)}")
print(f"testing accuracy of elastic net regression => {net.score(x_test, y_test)}")
```
