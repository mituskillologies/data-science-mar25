```python
# function

def outlier(column):

    q3 = np.percentile(column, 75)
    q1 = np.percentile(column, 25)

    iqr = q3 - q1

    ub = q3 + 1.5 * iqr
    lb = q1 - 1.5 * iqr

    return ub, lb
```

```python
outlier(data['Age'])
```

```python
outlier(data["EstimatedSalary"])
```

```python
ub, lb = outlier(data['Age'])
```

```python
# filter

data = data[(data["Age"] > lb) & (data["Age"] < ub)]
```
